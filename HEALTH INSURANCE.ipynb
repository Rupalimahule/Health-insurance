```python
for column in ['bmi', 'claim']:
    Q1 = df[column].quantile(0.35)
    Q3 = df[column].quantile(0.95)
    IQR = Q3 - Q1
    lower_bound = Q1 - 2.5 * IQR
    upper_bound = Q3 + 2.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column]> upper_bound)]

for column in ['bmi','claim']:   
  plt.figure(figsize=(10,5))
  sns.boxplot(x=data[column])
  plt.title(f'Boxplot After outlier treatment; {column}')
  plt.show()
```

```python
numeric_columns = data.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()
print(correlation_matrix)
plt.figure(figsize=(9,7))
sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm',fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
numerical_cols = ['age','weight','bmi','bloodpressure','claim']

for col in numerical_cols:
    plt.figure(figsize=(10,5))
    sns.histplot(data[col], kde=True,color='red')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='bmi',y ='claim', hue='smoker',data=data, palette='viridis')
plt.title('BHI vs Claim (smoker)')
plt.xlabel('BHI (BHI)')
plt.ylabel('Claim value')
plt.show()

plt.figure(figsize=(10,5))
sns.scatterplot(x='smoker',y ='claim',data=data)
plt.title('Claim Distribution by smoker')
plt.xlabel('smoker(0=No,1=Yes)')
plt.ylabel('Claim value')
plt.show()

```

```python
cross_tab = pd.crosstab(data['smoker'],data['hereditary_diseases'])
print(cross_tab)
cross_tab.plot(kind='bar',figsize=(10,5), stacked=True,colormap='viridis')
plt.title('Smoker Status and Hereditary Diseases')
plt.xlabel('Smoker(0=No,1=Yes)')
plt.ylabel('count')
plt.show()
```
